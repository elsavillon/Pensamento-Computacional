# Comparativo em despesas lançadas na API da Câmara dos Deputados entre os quatro deputados federais mais votados em 2022 - Guilherme Boulos (PSOL), Carla Zambelli (PL), Eduardo Bolsonaro (PL) e Ricardo Salles (PL)

#Função para acessar páginas de despesas

```python
def obter_despesas_candidato(candidato, legislatura):
    # Lista para guardar despesas
    despesas_candidato = []

    # Parâmetros da API (definição)
    headers = {'accept': 'application/json',}
    params = {'idLegislatura': legislatura, 'ordem': 'ASC', 'ordenarPor': 'ano',}

    # Primeira iteração
    response = requests.get(f'https://dadosabertos.camara.leg.br/api/v2/deputados/{candidato}/despesas', params=params, headers=headers)
    despesas = response.json()

    for despesa in despesas["dados"]:
        despesas_candidato.append(despesa)

    # Pegar todas as outras, sem saber ao certo o número de páginas (paginação "infinita")
    tem_proxima = True
    while tem_proxima:
        proxima = despesas['links'][1]['href']
        print(proxima)
        resposta = requests.get(proxima)
        despesas = resposta.json()
        for despesa in despesas["dados"]:
            despesas_candidato.append(despesa)

        if despesas['links'][1]['rel'] != 'next':
            tem_proxima = False

    return despesas_candidato
```

#Boulos

```python
deputado='Guilherme Boulos'
headers = {'accept': 'application/json',}
params = {'nome':deputado,'ordem': 'ASC','ordenarPor': 'nome',}
response_1 = requests.get('https://dadosabertos.camara.leg.br/api/v2/deputados', params=params, headers=headers)
```

```python
boulos_json = response_1.json()
```

```python
boulos_json
```

```python
boulos= 220639
legislatura_atual = 57

despesas_boulos= obter_despesas_candidato(boulos, legislatura_atual)
print(f'Despesas de {boulos}: {len(despesas_boulos)}')
```

```python
boulos=pd.DataFrame(despesas_boulos)
```

```python
boulos.to_csv('despesas_boulos.csv', index=False)
```
